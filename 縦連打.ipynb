```python

import os

# ファイル名変更
old_filename = 'LOST LOVE(ANOTHER).bme'
new_filename = 'LOST LOVE(ANOTHER).txt'

# ファイルの拡張子を変更
os.rename(old_filename, new_filename)

# ファイルを'r+'モードで開く
f = open(new_filename, 'r+', encoding='UTF-8')

# ファイルの内容をすべて読み込む
dtl = f.readlines()

# 数字チェック用のヘルパー関数
def is_numeric(s):
    return s.isdigit()

# 各行の末尾から改行文字を削除し、空文字列を'00'に変換
dtl = [line.rstrip('\n') if line.rstrip('\n') != '' else '00' for line in dtl]

# MAIN DATA FIELDの行を探す
main_data_start = False
main_data = []

for line in dtl:
    if main_data_start:
        main_data.append(line)
    if "MAIN DATA FIELD" in line:
        main_data_start = True

# 処理用の二次元配列
dtlb = []
current_group = []

# 4文字目が変わるときにグループを分けるための変数
previous_char4 = None

# 6文字目の値に基づいて初期化された辞書を作成
char6_values = '61234589'
current_group_dict = {char: '0' for char in char6_values}

for element in main_data:
    if len(element) > 7 and is_numeric(element[1:6]) and element[4] != '0':  # 2文字目から6文字が数字かどうかを確認し、5文字目が'0'でないことを確認
        char4 = element[3]  # 4文字目
        char6 = element[5]  # 6文字目

        if previous_char4 is not None and char4 != previous_char4:
            # すべてのchar6_valuesが存在するか確認し、存在しない場合は'0'を追加
            for char in char6_values:
                if current_group_dict[char] == '0':
                    current_group.append('0')
                else:
                    current_group.append(current_group_dict[char])
            
            dtlb.append(current_group)
            current_group = []
            current_group_dict = {char: '0' for char in char6_values}  # 新しいグループ用に初期化
        
        previous_char4 = char4

        result = ''  # 結果を一時的に格納する文字列
        for i in range(7, len(element), 2):  # 8文字目以降から2文字ずつ取り出す
            if element[i:i+2] == '00':
                result += '0'
            else:
                result += '1'
        
        current_group_dict[char6] = result

# 最後のグループの処理
for char in char6_values:
    if current_group_dict[char] == '0':
        current_group.append('0')
    else:
        current_group.append(current_group_dict[char])

if current_group:
    dtlb.append(current_group)

# 結果を表示
print(dtlb)

# ファイルを閉じる
f.close()

# 元のファイル名に戻す（必要に応じて）
os.rename(new_filename, old_filename)


# 16文字以上あり、1が連続する要素の番号を表示する
for index, group in enumerate(dtlb):
    for item in group:
        if len(item) >= 16 and'1' * 6 in item:  # 1が6つ連続する部分文字列が存在するかどうかをチェック
            print(f"Group {index + 1}, Item: {item}")


```
